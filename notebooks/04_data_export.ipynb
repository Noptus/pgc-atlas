# 04 - Data Export

Filter and export PGC GWAS data in multiple formats.


```python
from pgc_explorer.data_loader import PGCDataLoader
from pgc_explorer.export import export_data

loader = PGCDataLoader()
```

```python
# Load significant hits for schizophrenia
df = loader.load_top_hits("schizophrenia", n=10_000)
print(f"Total variants loaded: {len(df)}")
```

## Export to CSV


```python
csv_bytes = export_data(df, fmt='csv', significant_only=True)
with open('scz_significant.csv', 'wb') as f:
    f.write(csv_bytes)
print(f"Exported {len(csv_bytes):,} bytes to scz_significant.csv")
```

## Export to BED Format

Useful for loading into genome browsers like UCSC or IGV:


```python
bed_bytes = export_data(df, fmt='bed', significant_only=True)
with open('scz_hits.bed', 'wb') as f:
    f.write(bed_bytes)
print(f"BED file: {len(bed_bytes):,} bytes")
```

## Export to Parquet

Efficient columnar format for downstream analysis:


```python
parquet_bytes = export_data(df, fmt='parquet', columns=['snp', 'chr', 'bp', 'pval', 'effect'])
with open('scz_summary.parquet', 'wb') as f:
    f.write(parquet_bytes)
print(f"Parquet file: {len(parquet_bytes):,} bytes")
```
