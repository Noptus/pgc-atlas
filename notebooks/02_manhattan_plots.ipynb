# 02 - Manhattan Plots

Create interactive Manhattan plots for any psychiatric disorder in the PGC collection.


```python
from pgc_explorer.data_loader import PGCDataLoader
from pgc_explorer.manhattan import prepare_manhattan_data
import plotly.graph_objects as go

loader = PGCDataLoader()
```

```python
# Load schizophrenia data and generate Manhattan plot
df = loader.load_top_hits("schizophrenia", n=50_000)
plot_data = prepare_manhattan_data(df)

fig = go.Figure(data=plot_data['traces'], layout=plot_data['layout'])
fig.update_layout(
    title='Schizophrenia GWAS - Manhattan Plot',
    height=500, width=1200,
    template='plotly_dark',
)
fig.show()
```

## Manhattan Plots for All Disorders


```python
from pgc_explorer.config import PGC_DATASETS

for disorder, meta in PGC_DATASETS.items():
    try:
        df = loader.load_top_hits(disorder, n=20_000)
        if df is not None and not df.empty:
            plot_data = prepare_manhattan_data(df)
            fig = go.Figure(data=plot_data['traces'], layout=plot_data['layout'])
            fig.update_layout(title=f"{meta.display_name} GWAS", height=300, width=1200, template='plotly_dark')
            fig.show()
    except Exception as e:
        print(f"Skipping {disorder}: {e}")
```

## QQ Plot


```python
from pgc_explorer.qq_plot import prepare_qq_data

df = loader.load_top_hits("schizophrenia", n=50_000)
qq = prepare_qq_data(df)

fig = go.Figure(data=qq['traces'], layout=qq['layout'])
fig.update_layout(title='Schizophrenia QQ Plot', height=500, width=500, template='plotly_dark')
fig.show()

print(f"Genomic inflation factor: lambda_GC = {qq['lambda_gc']:.4f}")
```
