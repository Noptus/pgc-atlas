# 03 - Cross-Disorder Analysis

Explore shared genetic architecture between psychiatric disorders.


```python
from pgc_explorer.data_loader import PGCDataLoader
from pgc_explorer.cross_trait import cross_trait_matrix, compute_sign_concordance
import plotly.graph_objects as go
import plotly.express as px

loader = PGCDataLoader()
```

## Load Data for Multiple Disorders


```python
disorders_to_compare = ['schizophrenia', 'bipolar', 'mdd', 'adhd', 'autism', 'ptsd']

datasets = {}
for d in disorders_to_compare:
    try:
        df = loader.load_top_hits(d, n=20_000)
        if df is not None and not df.empty:
            datasets[d] = df
            print(f"Loaded {d}: {len(df)} variants")
    except Exception as e:
        print(f"Failed to load {d}: {e}")
```

## Genetic Correlation Matrix


```python
if len(datasets) >= 2:
    corr_matrix = cross_trait_matrix(datasets, method='correlation')
    
    fig = px.imshow(
        corr_matrix,
        color_continuous_scale='RdBu_r',
        zmin=-0.5, zmax=0.5,
        title='Cross-Disorder Genetic Correlation',
    )
    fig.update_layout(height=600, width=700, template='plotly_dark')
    fig.show()
```

## Pairwise Sign Concordance

Check if shared significant SNPs have the same direction of effect:


```python
if 'schizophrenia' in datasets and 'bipolar' in datasets:
    result = compute_sign_concordance(datasets['schizophrenia'], datasets['bipolar'])
    print(f"SCZ vs BIP:")
    print(f"  Shared significant SNPs: {result['n_shared']}")
    print(f"  Sign concordance: {result['concordance']:.1%}")
    print(f"  Binomial P-value: {result['pvalue']:.2e}")
```
